```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# derivatives
x = np.linspace(-5, 5, 1000)
sigmoid = 1/(1+np.exp(-x))
d_sigmoid = sigmoid * (1-sigmoid)

tanh = np.tanh(x)
d_tanh = 1 - tanh**2

relu = (x>=0)*x
d_relu = (x>=0)

fig, ax = plt.subplots()
ax.plot(x, d_sigmoid)
ax.plot(x, d_tanh)
ax.plot(x, d_relu)
ax.grid()

ax.legend(['derivative of sigmoid', 'derivative of tanh', 'derivative of relu'])

fig.savefig("derivatives.png")
plt.show()
```

```python
# functions
x = np.linspace(-2.5, 2.5, 1000)
sigmoid = 1/(1+np.exp(-x))

tanh = np.tanh(x)

relu = (x>=0)*x

fig, ax = plt.subplots()
ax.plot(x, sigmoid)
ax.plot(x, tanh)
ax.plot(x, relu)
ax.grid()

ax.legend(['sigmoid', 'tanh', 'relu'])
plt.xlabel('x')
plt.ylabel('f(x)')

fig.savefig("activations.png")
plt.show()
```
