```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid

x = np.linspace(-5, 5, 1000)
sigmoid = 1/(1+np.exp(-x))
d_sigmoid = sigmoid * (1-sigmoid)

fig, ax = plt.subplots()
ax.plot(x, sigmoid)
ax.plot(x, d_sigmoid)
ax.legend(['sigmoid activation','derivative of sigmoid'])
ax.grid()


fig.savefig("sigmoid.png")
plt.show()
```

```python
# tanh
x = np.linspace(-5, 5, 1000)
tanh = np.tanh(x)
d_tanh = 1 - tanh**2

fig, ax = plt.subplots()
ax.plot(x, tanh)
ax.plot(x, d_tanh)
ax.legend(['tanh activation','derivative of tanh'])
ax.grid()

fig.savefig("tanh.png")
plt.show()
```

```python
# relu
x = np.linspace(-2, 2, 1000)
tanh = (x>=0)*x
d_tanh = x>=0

fig, ax = plt.subplots()
ax.plot(x, tanh)
ax.plot(x, d_tanh)
ax.legend(['relu activation','derivative of relu'])
ax.grid()

fig.savefig("relu.png")
plt.show()
```

```python
# leaky relu
x = np.linspace(-2, 2, 1000)
leaky_relu = np.where(x >= 0, x, x * 0.1)
d_leaky_relu = np.where(x >= 0, 1, 0.1)

fig, ax = plt.subplots()
ax.plot(x, leaky_relu)
ax.plot(x, d_leaky_relu)

ax.legend(['leaky relu activation','derivative of leaky relu'])
ax.grid()

fig.savefig("leaky_relu.png")
plt.show()
```

```python
# elu
x = np.linspace(-5, 5, 1000)
elu = np.where(x > 0, x, 0.9*(np.exp(x)-1))
d_elu = np.where(x > 0, 1, 0.9*np.exp(x))

fig, ax = plt.subplots()
ax.plot(x, elu)
ax.plot(x, d_elu)

ax.legend(['elu activation','derivative of elu'])
ax.grid()

fig.savefig("elu.png")
plt.show()
```
