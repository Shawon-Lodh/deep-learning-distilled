```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,8,10)
x = np.sort(x)

T = [1,2,5]

for t in T:
    m = np.max(x)
    s = np.sum(np.exp((x-m)/t))
    y = np.exp((x-m)/t) / s
    plt.plot(x,y)

plt.legend(['T=1', 'T=2', 'T=5'])
plt.savefig('softmax_temperature.png')
plt.show()
```
